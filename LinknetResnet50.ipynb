```python
from google.colab import drive
import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2
from xml.etree import ElementTree as ET

# Mount Google Drive
drive.mount('/content/drive')

# Wayanad reference point
ref_lat = 11.6577805556
ref_lon = 76.1117222222

# Haversine formula to calculate distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Load the dataset from Google Drive
excel_path = '/content/drive/MyDrive/converted_coordinates_dms_india_with_dates_2024.xlsx'
df = pd.read_excel(excel_path)

# Calculate distances and filter coordinates within 100 km
df['distance'] = df.apply(lambda row: haversine(ref_lat, ref_lon, row['lat_dd'], row['lon_dd']), axis=1)
filtered_df = df[df['distance'] <= 100]

# Create KML file
kml = ET.Element('kml', xmlns="http://www.opengis.net/kml/2.2")
doc = ET.SubElement(kml, 'Document')
ET.SubElement(doc, 'name').text = 'Coordinates within 100 km of Wayanad'

for index, row in filtered_df.iterrows():
    placemark = ET.SubElement(doc, 'Placemark')
    ET.SubElement(placemark, 'name').text = f"Point {index + 1} - {row['valid_time']}"
    point = ET.SubElement(placemark, 'Point')
    ET.SubElement(point, 'coordinates').text = f"{row['lon_dd']},{row['lat_dd']},0"

# Save KML to Google Drive
kml_string = ET.tostring(kml, encoding='unicode', method='xml')
kml_path = '/content/drive/MyDrive/wayanad_coordinates.kml'
with open(kml_path, 'w', encoding='utf-8') as f:
    f.write('<?xml version="1.0" encoding="UTF-8"?>\n')
    f.write(kml_string)

print(f"KML file saved to {kml_path} with {len(filtered_df)} coordinates within 100 km of Wayanad.")
```

```python
import torch
import torchvision.models as models
import torch.nn as nn

# ✅ Define LinkNet with ResNet50
class LinkNetResNet50(nn.Module):
    def __init__(self, num_classes=1):
        super(LinkNetResNet50, self).__init__()

        resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)

        self.encoder1 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu, resnet.maxpool)
        self.encoder2 = resnet.layer1
        self.encoder3 = resnet.layer2
        self.encoder4 = resnet.layer3
        self.encoder5 = resnet.layer4

        self.decoder5 = self._decoder_block(2048, 1024)
        self.decoder4 = self._decoder_block(1024, 512)
        self.decoder3 = self._decoder_block(512, 256)
        self.decoder2 = self._decoder_block(256, 64)
        self.decoder1 = self._decoder_block(64, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _decoder_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, in_channels // 4, kernel_size=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, in_channels // 4, kernel_size=3, padding=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, out_channels, kernel_size=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        e1 = self.encoder1(x)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        e5 = self.encoder5(e4)

        d5 = self.decoder5(nn.functional.interpolate(e5, size=e4.shape[-2:], mode="bilinear", align_corners=False)) + e4
        d4 = self.decoder4(nn.functional.interpolate(d5, size=e3.shape[-2:], mode="bilinear", align_corners=False)) + e3
        d3 = self.decoder3(nn.functional.interpolate(d4, size=e2.shape[-2:], mode="bilinear", align_corners=False)) + e2
        d2 = self.decoder2(nn.functional.interpolate(d3, size=e1.shape[-2:], mode="bilinear", align_corners=False)) + e1
        d1 = self.decoder1(nn.functional.interpolate(d2, size=x.shape[-2:], mode="bilinear", align_corners=False))

        out = self.final_conv(d1)
        return out

# ✅ Load Model Weights
device = "cuda" if torch.cuda.is_available() else "cpu"
model = LinkNetResNet50(num_classes=1).to(device)
model.load_state_dict(torch.load("/content/drive/MyDrive/LandslideDataset/LinkNet_ResNet50_v2.pth", map_location=device))
print("✅ Retrained model loaded successfully!")
```

```python
import os
import numpy as np
import tifffile
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision import transforms
from sklearn.metrics import precision_score, recall_score, f1_score
from google.colab import drive
import cv2
from tqdm import tqdm
import gc

# Mount Google Drive
drive.mount('/content/drive')

# Define the model architecture
class LinkNetResNet50(nn.Module):
    def __init__(self, num_classes=1):
        super(LinkNetResNet50, self).__init__()
        resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        self.encoder1 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu, resnet.maxpool)
        self.encoder2 = resnet.layer1
        self.encoder3 = resnet.layer2
        self.encoder4 = resnet.layer3
        self.encoder5 = resnet.layer4
        self.decoder5 = self._decoder_block(2048, 1024)
        self.decoder4 = self._decoder_block(1024, 512)
        self.decoder3 = self._decoder_block(512, 256)
        self.decoder2 = self._decoder_block(256, 64)
        self.decoder1 = self._decoder_block(64, 64)
        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _decoder_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, in_channels // 4, kernel_size=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, in_channels // 4, kernel_size=3, padding=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, out_channels, kernel_size=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        e1 = self.encoder1(x)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        e5 = self.encoder5(e4)
        d5 = self.decoder5(nn.functional.interpolate(e5, size=e4.shape[-2:], mode="bilinear", align_corners=False)) + e4
        d4 = self.decoder4(nn.functional.interpolate(d5, size=e3.shape[-2:], mode="bilinear", align_corners=False)) + e3
        d3 = self.decoder3(nn.functional.interpolate(d4, size=e2.shape[-2:], mode="bilinear", align_corners=False)) + e2
        d2 = self.decoder2(nn.functional.interpolate(d3, size=e1.shape[-2:], mode="bilinear", align_corners=False)) + e1
        d1 = self.decoder1(nn.functional.interpolate(d2, size=x.shape[-2:], mode="bilinear", align_corners=False))
        out = self.final_conv(d1)
        return out

# Paths (update these to your actual paths in Google Drive)
image_folder = '/content/drive/MyDrive/matched_test_images'
mask_folder = '/content/drive/MyDrive/matched_test_masks'
model_path = '/content/drive/MyDrive/LandslideDataset/LinkNet_ResNet50_v2.pth'
output_folder = '/content/drive/MyDrive/predicted_masks/'

# Create output folder if it doesn’t exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Tile parameters
tile_size = 256
overlap = 32

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the model
model = LinkNetResNet50(num_classes=1).to(device)
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()
print("✅ Retrained model loaded successfully!")

# Functions for tile extraction and reconstruction
def extract_tiles(image, tile_size, overlap):
    h, w = image.shape[:2]
    tiles = []
    positions = []
    step = tile_size - overlap
    for y in range(0, h, step):
        for x in range(0, w, step):
            y_end = min(y + tile_size, h)
            x_end = min(x + tile_size, w)
            y_start = y_end - tile_size
            x_start = x_end - tile_size
            tile = image[y_start:y_end, x_start:x_end]
            if tile.shape[:2] != (tile_size, tile_size):
                tile = cv2.copyMakeBorder(tile, 0, tile_size - tile.shape[0], 0, tile_size - tile.shape[1], cv2.BORDER_CONSTANT, value=0)
            tiles.append(tile)
            positions.append((y_start, x_start, y_end, x_end))
    return tiles, positions

def reconstruct_image(tiles, positions, image_shape, tile_size, overlap):
    h, w = image_shape[:2]
    reconstructed = np.zeros((h, w), dtype=np.float32)
    count = np.zeros((h, w), dtype=np.float32)
    for tile, (y_start, x_start, y_end, x_end) in zip(tiles, positions):
        tile_h, tile_w = tile.shape[:2]
        valid_h = min(tile_size, y_end - y_start)
        valid_w = min(tile_size, x_end - x_start)
        tile = tile[:valid_h, :valid_w]
        reconstructed[y_start:y_end, x_start:x_end] += tile
        count[y_start:y_end, x_start:x_end] += 1
    reconstructed = np.divide(reconstructed, count, where=count != 0)
    return reconstructed

# Get list of image and mask files
image_files = sorted([f for f in os.listdir(image_folder) if f.endswith(('.tiff', '.tif'))])
mask_files = sorted([f for f in os.listdir(mask_folder) if f.endswith(('.tiff', '.tif'))])

# Validate matching files
if len(image_files) != len(mask_files):
    raise ValueError("Number of images and masks do not match.")

# Transform for model input
transform = transforms.ToTensor()

# Store metrics
metrics = {'accuracy': [], 'precision': [], 'recall': [], 'f1': [], 'iou': []}

# Process each image-mask pair
for img_file, mask_file in tqdm(zip(image_files, mask_files), total=len(image_files), desc="Evaluating"):
    # Load image and mask
    image = tifffile.imread(os.path.join(image_folder, img_file))
    mask = tifffile.imread(os.path.join(mask_folder, mask_file))

    # Preprocess mask to binary
    mask = (mask > 0).astype(np.uint8)

    # Normalize image
    image = image.astype(np.float32) / 255.0

    # Extract tiles
    tiles, positions = extract_tiles(image, tile_size, overlap)

    # Process tiles
    predicted_tiles = []
    with torch.no_grad():
        for tile in tiles:
            if len(tile.shape) == 2:  # Grayscale
                tile = np.expand_dims(tile, axis=2)
            tile_tensor = transform(tile)
            if tile_tensor.shape[0] == 1:  # Single channel
                tile_tensor = tile_tensor.repeat(3, 1, 1)
            tile_tensor = tile_tensor.unsqueeze(0).to(device)
            pred = model(tile_tensor)
            pred = torch.sigmoid(pred).squeeze().cpu().numpy()
            pred = (pred > 0.5).astype(np.uint8)
            predicted_tiles.append(pred)

    # Reconstruct predicted mask
    predicted_mask = reconstruct_image(predicted_tiles, positions, image.shape, tile_size, overlap)
    predicted_mask = (predicted_mask > 0.5).astype(np.uint8)

    # Save predicted mask
    tifffile.imwrite(os.path.join(output_folder, f"pred_{img_file}"), predicted_mask)

    # Calculate metrics
    true_flat = mask.flatten()
    pred_flat = predicted_mask.flatten()
    TP = np.sum((pred_flat == 1) & (true_flat == 1))
    TN = np.sum((pred_flat == 0) & (true_flat == 0))
    FP = np.sum((pred_flat == 1) & (true_flat == 0))
    FN = np.sum((pred_flat == 0) & (true_flat == 1))
    accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) > 0 else 0
    precision = precision_score(true_flat, pred_flat, zero_division=0)
    recall = recall_score(true_flat, pred_flat, zero_division=0)
    f1 = f1_score(true_flat, pred_flat, zero_division=0)
    iou = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0

    # Store metrics
    metrics['accuracy'].append(accuracy)
    metrics['precision'].append(precision)
    metrics['recall'].append(recall)
    metrics['f1'].append(f1)
    metrics['iou'].append(iou)

    # Print per-image metrics
    print(f"\nMetrics for {img_file}:")
    print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}, IoU: {iou:.4f}")

    # Clear memory
    del image, mask, tiles, predicted_tiles, predicted_mask
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

# Compute and print average metrics
avg_metrics = {key: np.mean(values) for key, values in metrics.items()}
print("\nAverage Metrics Across All Images:")
for key, value in avg_metrics.items():
    print(f"{key.capitalize()}: {value:.4f}")
```

```python
import cv2
import numpy as np
import torch
from torchvision import transforms

def load_and_split_image(image_path, tile_size=(256, 256)):
    """
    Load an image and split it into tiles of size (tile_size x tile_size).
    """
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Failed to load image: {image_path}")

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

    h, w, _ = image.shape
    tiles = []

    # Divide image into non-overlapping tiles
    for y in range(0, h, tile_size[1]):
        for x in range(0, w, tile_size[0]):
            tile = image[y:y + tile_size[1], x:x + tile_size[0]]
            if tile.shape[:2] == tile_size:  # Only keep full tiles
                tiles.append(tile)

    return tiles, (h, w), image  # Return tiles, original size, and full image

# Path to your uploaded test image
test_image_path = "/content/drive/MyDrive/landslide optical images/Noney_manipur_2022.jpg"
tiles, original_size, original_image = load_and_split_image(test_image_path)
print(f"Total tiles generated: {len(tiles)}")
```

```python
def preprocess_tiles(tiles, device):
    """
    Preprocess the tiles to match model input format.
    """
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    preprocessed_tiles = [transform(tile).unsqueeze(0).to(device) for tile in tiles]
    return preprocessed_tiles

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
preprocessed_tiles = preprocess_tiles(tiles, device)
```

```python
def predict_tiles(model, preprocessed_tiles):
    """
    Run model prediction on each tile.
    """
    model.eval()
    predicted_masks = []

    with torch.no_grad():
        for tile in preprocessed_tiles:
            output = model(tile)
            output = torch.sigmoid(output)  # Apply sigmoid activation
            output = output.squeeze().cpu().numpy()
            predicted_masks.append(output)

    return predicted_masks

# Assuming you have already loaded your trained model as 'model'
predicted_masks = predict_tiles(model, preprocessed_tiles)
```

```python
def reconstruct_mask(predicted_masks, original_size, tile_size=(256, 256)):
    """
    Reconstruct the final segmentation mask from the predicted tiles.
    """
    h, w = original_size
    mask = np.zeros((h, w))

    index = 0
    for y in range(0, h, tile_size[1]):
        for x in range(0, w, tile_size[0]):
            if index < len(predicted_masks):
                mask[y:y + tile_size[1], x:x + tile_size[0]] = predicted_masks[index]
                index += 1

    return mask

full_predicted_mask = reconstruct_mask(predicted_masks, original_size)
```

```python
import matplotlib.pyplot as plt

def visualize_results(original_image, full_predicted_mask):
    """
    Display the original image and the reconstructed predicted mask.
    """
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(full_predicted_mask, cmap="gray")
    plt.title("Predicted Mask")
    plt.axis("off")

    plt.show()

visualize_results(original_image, full_predicted_mask)
```

```python
import torch
import torchvision.models as models
import torch.nn as nn

# ✅ Define LinkNet with ResNet50
class LinkNetResNet50(nn.Module):
    def __init__(self, num_classes=1):
        super(LinkNetResNet50, self).__init__()

        resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)

        self.encoder1 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu, resnet.maxpool)
        self.encoder2 = resnet.layer1
        self.encoder3 = resnet.layer2
        self.encoder4 = resnet.layer3
        self.encoder5 = resnet.layer4

        self.decoder5 = self._decoder_block(2048, 1024)
        self.decoder4 = self._decoder_block(1024, 512)
        self.decoder3 = self._decoder_block(512, 256)
        self.decoder2 = self._decoder_block(256, 64)
        self.decoder1 = self._decoder_block(64, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _decoder_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, in_channels // 4, kernel_size=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, in_channels // 4, kernel_size=3, padding=1),
            nn.BatchNorm2d(in_channels // 4),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 4, out_channels, kernel_size=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        e1 = self.encoder1(x)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        e5 = self.encoder5(e4)

        d5 = self.decoder5(nn.functional.interpolate(e5, size=e4.shape[-2:], mode="bilinear", align_corners=False)) + e4
        d4 = self.decoder4(nn.functional.interpolate(d5, size=e3.shape[-2:], mode="bilinear", align_corners=False)) + e3
        d3 = self.decoder3(nn.functional.interpolate(d4, size=e2.shape[-2:], mode="bilinear", align_corners=False)) + e2
        d2 = self.decoder2(nn.functional.interpolate(d3, size=e1.shape[-2:], mode="bilinear", align_corners=False)) + e1
        d1 = self.decoder1(nn.functional.interpolate(d2, size=x.shape[-2:], mode="bilinear", align_corners=False))

        out = self.final_conv(d1)
        return out

# ✅ Load Model Weights
device = "cuda" if torch.cuda.is_available() else "cpu"
model = LinkNetResNet50(num_classes=1).to(device)
model.load_state_dict(torch.load("/content/drive/MyDrive/LandslideDataset/LinkNet_ResNet50_v2.pth", map_location=device))
print("✅ Retrained model loaded successfully!")
```
